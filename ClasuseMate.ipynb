```python
from langchain.document_loaders import PyPDFLoader

#
loader = PyPDFLoader("BAJHLIP23020V012223.pdf")
documents = loader.load()
#
print(documents)
```

```python
pages=[]
for d in documents:
  completion = client.chat.completions.create(
  model="qwen/qwen3-30b-a3b:free",
  messages=[
    {
      "role": "user",
      "content": [
        {
          "type": "text",
          "text": f"message = {d.page_content}.  remove the header and unnecessary information and just give the policy text from the given message"
        },
      ]
    }
  ]
  )
  pages.append(completion.choices[0].message.content)

```

```python
from langchain_community.embeddings.fastembed import FastEmbedEmbeddings
embed_model = FastEmbedEmbeddings(model_name="BAAI/bge-base-en-v1.5")
```

```python
from langchain_experimental.text_splitter import SemanticChunker
from langchain_openai.embeddings import OpenAIEmbeddings

semantic_chunker = SemanticChunker(embed_model, breakpoint_threshold_type="percentile")
#
semantic_chunks = semantic_chunker.create_documents([d.page_content for d in documents])

for i, chunk in enumerate(semantic_chunks):
    chunk.metadata["source"] = "policy.pdf"
    chunk.metadata["page"] = i + 1  # or actual page number if tracked

```

```python
for i, chunk in enumerate(semantic_chunks):
    print(f"\n--- Chunk {i + 1} ---")
    print(chunk.page_content)
    print("Length:", len(chunk.page_content))
```
